```python
# Install library openpyxl jika belum terpasang
!pip install openpyxl

import openpyxl
from google.colab import files

# Unggah file Excel
uploaded = files.upload()

# Ambil nama file yang diunggah
file_path = list(uploaded.keys())[0]

# Load workbook
workbook = openpyxl.load_workbook(file_path)
sheet = workbook.active

# Kode warna kuning yang ditemukan (FFFFFF00)
yellow_cells_text = []

for row in sheet.iter_rows():
    for cell in row:
        # Cek jika sel memiliki fill dan fill color RGB-nya kuning (FFFFFF00)
        if cell.fill.start_color.index == 'FFFFFF00':
            yellow_cells_text.append(cell.value)

# Menampilkan hasil
print("Teks dari sel dengan background kuning:")
for text in yellow_cells_text:
    print(text)
```
