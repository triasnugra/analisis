```python
# Deskriptif statistik
descriptive_stats = df.describe()
print(descriptive_stats)
```

```python
# Visualisasi dengan histogram
df.plot(kind='hist', bins=5, alpha=0.7, figsize=(10, 6))
plt.title("Distribusi Skala Likert untuk Semua Kategori")
plt.xlabel("Skala Likert")
plt.ylabel("Jumlah Responden")
plt.legend(df.columns)
plt.show()

# Visualisasi dengan boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Distribusi Skala Likert - Boxplot")
plt.ylabel("Skala Likert")
plt.show()
```

```python
# Menghitung frekuensi setiap kategori (1-5) untuk setiap kolom
frequency = df.apply(lambda x: x.value_counts().sort_index(), axis=0)
print(frequency)
```

```python
# Korelasi antar kategori
correlation = df.corr()
print(correlation)

# Visualisasi korelasi dengan heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Korelasi Antarkategori")
plt.show()
```
