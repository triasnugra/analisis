```python
import requests
from bs4 import BeautifulSoup
import time

# Function to scrape publication titles and years from a given Google Scholar profile URL
def scrape_google_scholar(url):
    all_data = []
    page_num = 0  # Initialize the page number

    while len(all_data) < 40:  # Continue until we have at least 40 titles
        # Construct the URL for the specific page
        page_url = f"{url}&cstart={page_num * 20}&view_op=list_works&sortby=pubdate"

        # Custom headers to mimic a browser
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36'
        }

        # Send a GET request to the URL with headers
        response = requests.get(page_url, headers=headers)

        # Check for any errors
        if response.status_code == 200:
            # Parse the page content
            soup = BeautifulSoup(response.text, 'html.parser')

            # Find all publication entries
            publications = soup.find_all('tr', class_='gsc_a_tr')

            # Check if publications are found on the page
            if not publications:
                print("No more publications found.")
                break

            # Extract titles and years
            for pub in publications:
                title = pub.find('a', class_='gsc_a_at').text
                year = pub.find('span', class_='gsc_a_h gsc_a_hc gs_ibl').text
                all_data.append((title, year))

            print(f"Fetched {len(all_data)} titles so far...")
        else:
            print(f"Failed to retrieve data: {response.status_code}")
            break

        # Increment page number for the next loop iteration
        page_num += 1

        # Add a delay to avoid triggering rate limits
        time.sleep(2)

    return all_data

# URL of the Google Scholar profile
scholar_url = "https://scholar.google.com/citations?hl=en&user=2sVgcfcAAAAJ"

# Scrape the data
data = scrape_google_scholar(scholar_url)

# Print the extracted data
for title, year in data:
    print(f"Title: {title}, Year: {year}")
```
