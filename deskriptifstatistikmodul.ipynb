```python
from google.colab import files
import pandas as pd

# UPLOAADDDD
uploaded = files.upload()

# READ YAAAAAAAAAAA
file_name = list(uploaded.keys())[0]
data = pd.read_excel(file_name)

# Show Data PLZ
data
```

```python
# 1. Frekuensi
print("\n1. Frekuensi:")
for column in data.columns:
    print(f"\nKolom: {column}")
    freq_table = data[column].value_counts().reset_index()
    freq_table.columns = [f"Nilai di '{column}'", "Frekuensi"]
    display(freq_table.style.background_gradient(cmap="coolwarm"))
```

```python
# 2. Statistik Deskriptif
print("\n2. Statistik Deskriptif:")
stat_summary = data.describe().transpose()
stat_summary['Skewness'] = data.skew()
stat_summary['Kurtosis'] = data.kurtosis()
display(stat_summary.style.background_gradient(cmap="viridis"))
```

```python
# 3. Distribusi Frekuensi
print("\n3. Distribusi Frekuensi (Numerik):")
for column in data.select_dtypes(include=np.number).columns:
    freq_dist = pd.cut(data[column], bins=5).value_counts().reset_index()
    freq_dist.columns = [f"Rentang '{column}'", "Frekuensi"]
    display(freq_dist.style.background_gradient(cmap="coolwarm"))
```

```python
# 4. Tabel Kontingensi
print("\n4. Tabel Kontingensi (Non-Numerik):")
non_numeric_columns = data.select_dtypes(exclude=np.number).columns
if len(non_numeric_columns) > 1:
    for i, col1 in enumerate(non_numeric_columns):
        for col2 in non_numeric_columns[i + 1:]:
            print(f"\nKontingensi antara {col1} dan {col2}:")
            crosstab = pd.crosstab(data[col1], data[col2])
            display(crosstab.style.background_gradient(cmap="YlGnBu"))
```

```python
# 6. Boxplot
print("\n6. Boxplot:")
for column in data.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=data[column], color='coral')
    plt.title(f"Boxplot: {column}")
    plt.xlabel(column)
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()
```
